```python
text = "summarize: The Inflation Reduction Act lowers prescription drug costs, health care costs, and energy costs. It's the most aggressive action on tackling the climate crisis in American history, which will lift up American workers and create good-paying, union jobs across the country. It'll lower the deficit and ask the ultra-wealthy and corporations to pay their fair share. And no one making under $400,000 per year will pay a penny more in taxes."
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="RishabhBot/my_awesome_billsum_model_summarize")
summarizer(text,max_length=50)
```

```python
def summarize_text(input_text, max_length=50):
    # Use the loaded summarization pipeline to generate summaries
    summary = summarizer(input_text, max_length=max_length)[0]['summary_text']
    return summary
```

```python
import gradio as gr

# Create an input component for text input
input_text = gr.Textbox(lines=10, label="Input Text")

# Create a button component for triggering the summarization
summarize_button = gr.Button(value="Summarize")

# Define the function that will be called when the button is clicked
def summarize(input_text):
    summary = summarize_text(input_text, max_length=50)
    return summary

# Create an interface with the input te xt box and summarize button
sum=gr.Interface(
    fn=summarize,
    inputs=input_text,
    outputs=gr.Textbox(label="Summary"),
    title="Text Summarization with Hugging Face Model",
    theme="default"
)

sum.launch(debug=True,share=True)
```
